```python
import cv2
import numpy as np
import math
import time
from PIL import ImageGrab as ig
import pyautogui

last_time = time.time()
while(True):
    #Detect Dinosaur
    img_rgb = np.array(ig.grab(bbox=(100,200,650,400)))
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) 
    template = cv2.imread('dino_template.jpg',0) 
    w, h = template.shape[::-1] 
    res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED) 
    threshold = 0.9
    loc = np.where( res >= threshold)
    
    dino_location = []
    # Draw a rectangle around the matched region. 
    for pt in zip(*loc[::-1]): 
        cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,255,255), 2)
        dino_location = [pt[0],pt[1],pt[0] + w, pt[1] + h]
    
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) 
    template = cv2.imread('obstacle_1_template.jpg',0) 
    w,h = template.shape[::-1] 
    res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED) 
    threshold = 0.65
    loc = np.where( res >= threshold) 
    
    
    detection_count = 0
    ground_obstacle_locations = []
    mask = np.zeros(img_rgb.shape[:2], np.uint8)
    for pt in zip(*loc[::-1]):
        if mask[math.floor(pt[1] + h/2), math.floor(pt[0] + w/2)] != 255:
            mask[pt[1]:pt[1]+h, pt[0]:pt[0]+w] = 255
            detection_count += 1
            cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,255,0), 1)
            obs_loc = [pt[0],pt[1],pt[0] + w, pt[1] + h]
            ground_obstacle_locations.append(obs_loc)
    # Show the final image with the matched area. 
    print("Detected {} obstacles".format(detection_count))
    
    
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) 
    template = cv2.imread('obstacle_2_template.jpg',0) 
    w,h = template.shape[::-1] 
    res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED) 
    threshold = 0.7
    loc = np.where( res >= threshold) 
    
    
    detection_count = 0
    mask = np.zeros(img_rgb.shape[:2], np.uint8)
    for pt in zip(*loc[::-1]):
        if mask[math.floor(pt[1] + h/2), math.floor(pt[0] + w/2)] != 255:
            mask[pt[1]:pt[1]+h, pt[0]:pt[0]+w] = 255
            detection_count += 1
            cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,255,0), 1)
            obs_loc = [pt[0], pt[1],pt[0] + w, pt[1] + h]
            ground_obstacle_locations.append(obs_loc)
    # Show the final image with the matched area. 
    print("Detected {} obstacles".format(detection_count))
    
    
    #play a move
    if len(dino_location)>0:
        dist_to_next_obstacle = 9999
        dino_x = dino_location[2]
        for obs in ground_obstacle_locations:
            if obs[0] > dino_x:
                dist_temp = obs[0]-dino_x
                if(dist_temp < dist_to_next_obstacle):
                    dist_to_next_obstacle = dist_temp
        if dist_to_next_obstacle < 100:
            pyautogui.press('up')
    cv2.imshow('Detected',img_rgb)
    if cv2.waitKey(25) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break
    print('Loop took {} seconds'.format(time.time()-last_time))
    last_time = time.time()
```
